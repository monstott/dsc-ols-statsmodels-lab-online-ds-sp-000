## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

data = pd.read_csv('Advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
data.head()
```

```python
# Get the 5-point statistics for data 
data.describe()
```

```python
# Describe the contents of this dataset

# This dataset contains 200 observaions.
# The dependent variable is 'sales'.
# The 3 independent variables are 'TV', 'radio', and 'newspaper'.
# These 3 predictors provide advertising budget dollars and the target variables provides sales dollars.
# From the mean, the largest budget belongs to 'TV'. 
# From the standard deviation, the most variable budget belongts to 'TV'.
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
for col in data:
    data[col].plot.hist(normed=True, label=col +' histogram')
    data[col].plot.kde(label = col +' kde')
    plt.legend()
    plt.show()
```

```python
# Record your observations here 
# None of the four variables in the dataset appear to have a completely normal distribution.
# There is right skew present in the 'newspaper' and 'sales' variables.
# 'TV' and 'radio' are fairly symmetrical but these variables also possess two peaks in their distributions.
# From these plots, analysis of skew and kurtosis variables is advised.
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(20, 8))
for idx, channel in enumerate(['TV', 'radio', 'newspaper']):
    data.plot(kind='scatter', x=channel, y='sales', ax=axs[idx], label=channel)
plt.legend()
plt.show()
```

```python
# Record yor observations on linearity here 
# From the scatterplots, 'TV' appears to have the most linear relationship with 'sales'.
# 'radio' may have a linear relationship with 'sales' but the variance in observations is very high.
# 'newspaper' does not appear to have a linear relationship with 'sales'. There is no pattern present.
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf

# build the formula 
f = 'sales~TV'

# create a fitted model in one line
model = smf.ols(formula=f, data=data).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
tv_minmax = pd.DataFrame({'TV': [data.TV.min(), data.TV.max()]})
print('TV min and max values:\n', tv_minmax.head())

# make predictions for those x values and store them
preds = model.predict(tv_minmax)
print('\nSales predictions:\n', preds)

# first, plot the observed data and the least squares line
data.plot(kind='scatter', x='TV', y='sales')
plt.plot(tv_minmax, preds)
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
# The Y and Fitted vs. X plot shows an increase in the distance between model and observation values as the value of 'TV' increases.
# This apperance of heteroscedasticity is confirmed in the Residuals versus TV plot which has an obvious cone shape that increases with 'TV'.
# Analysis of the Partial regressoin plot and CCPR Plot is not necessary since only one independent variable is present.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats

residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption
# Visual inspection of the Q-Q Plot indicates that the normality assumption is appropriate for the data.
# There is no extreme divergence from the red normal distribution line.
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here

# Model
f = 'sales~radio'
model = smf.ols(formula=f, data=data).fit()

# Prediction Line Plot
radio_minmax = pd.DataFrame({'radio': [data.radio.min(), data.radio.max()]});
preds = model.predict(radio_minmax)
data.plot(kind='scatter', x='radio', y='sales');
plt.plot(radio_minmax, preds);
plt.show()

# Heteroscedasticity Chek
fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, "radio", fig=fig)
plt.show()

# Q-Q Normality Check
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
model.summary()
```

```python
# Record your observations here for goodnes of fit 

# The prediction line plot for 'radio' shows a greater deviation of observations from the model than 'TV'.
# The linear assumption is not as strong for 'radio' as it is for 'TV'.
# The heteroscedasticity check shows that the residuals increase in value with 'radio'. This is similar to what was seen for 'TV'.
# The homoscedasticity assumption cannot be applied to the data.
# The normality check reveals higher skew and kurtosis for 'radio' than 'TV'. This is further confirmed in comparison of the model summaries.
# The normality assumption does not appear to hold.
# The'radio' model explains 33.2% of the variation in 'sales' compared to the 61.2% that is explained by the 'TV' model.
```
